### spectrogram

```python
BASE_PATH = '/content/drive/MyDrive/기계팀플/Data/genres_original/'
```

```python
genre = "blues"

plt.rcParams["figure.figsize"] = (10,8)
fig, ax = plt.subplots(nrows=5, ncols=1, sharex=True)

for i in range(5) :
  rand_num = random.randint(0, 99)

  filename = genre + "/{}.{:05d}.wav".format(genre, rand_num)
  print(filename)

  data_, sampling_rate = librosa.load(BASE_PATH + filename, sr=sr, offset=0.0, duration=30)

  audio = librosa.amplitude_to_db(np.abs(librosa.stft(data_)), ref=np.max)
  librosa.display.specshow(audio, y_axis='linear', x_axis='time', sr=sr, ax=ax[i])
```

## waveplot - (Harmonic + Percussive)
고조파/타악기 구성 요소

```python
genre = "pop"

plt.rcParams["figure.figsize"] = (10,4)

filename = genre + "/{}.{:05d}.wav".format(genre, 1)
print(filename)

data_, sampling_rate = librosa.load(BASE_PATH + filename, sr=sr, offset=0.0, duration=30)
audio, _ = librosa.effects.trim(data_)
y_harm, y_perc = librosa.effects.hpss(audio)

librosa.display.waveplot(y_harm, sr=sr, alpha=0.25)
librosa.display.waveplot(y_perc, sr=sr, color='b', alpha=0.5)

plt.title('Harmonic + Percussive')
plt.tight_layout()
plt.show()
```
